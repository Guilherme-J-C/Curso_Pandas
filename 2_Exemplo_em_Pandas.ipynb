```python
# Importando biblioteca pandas no Python
import pandas as pd
```

```python
combustiveis_df = pd.read_excel("ca-2021-02.xlsx")
display(combustiveis_df)
```

```python
# Inserção simples de dado
combustiveis_df['Ativo'] = True
display(combustiveis_df.head())
```

```python
# Preencher uma coluna 'Valor de Venda - Status' que verifica o seguinte:
# se o valor de venda for maior que 6,5 reais, ele fala que tá Caro..caso contrário, está Barato
import numpy as np

combustiveis_df['Status do Valor de Venda'] = np.where(combustiveis_df['Valor de Venda'] > 6.5, 'Caro', 'Barato')
display(combustiveis_df[['Revenda', 'Valor de Venda', 'Status do Valor de Venda']])
```

```python
# E se quiser uma coluna nova chamada "Obs" que preencha automaticamente "Município onde o professor mora."
# quando o Município for INDAIATUBA em SP ...
combustiveis_df['Obs'] = np.where(combustiveis_df['Municipio'] == 'INDAIATUBA', 'Município onde o professor mora.', '')
display(combustiveis_df.loc[combustiveis_df['Municipio'].isin(['INDAIATUBA', 'CAMPINAS']), ['Municipio', 'Obs']])
```

```python
# Calcular postos de galolina por habitante temos na amostragem de combustiveis nov/2021
num_habitantes_df = pd.read_csv('ibge_num_habitantes_estimado.csv', sep=";")
num_habitantes_df.rename(columns={"Estado": "Estado - Sigla"}, inplace=True)
display(num_habitantes_df)
```

```python
#  Fazer um MERGE dos dous dataframes
coluna = ['Municipio', 'Estado - Sigla']
merge_df = combustiveis_df.merge(num_habitantes_df, how="inner", on=coluna)
display(merge_df)
print(merge_df.info())
```

```python
# Destruir coluna vazia
merge_df.dropna(axis=1, inplace=True)
print(merge_df.info())
```

```python
# Destruir colunas específicas
colunas = ['Regiao - Sigla', 'Nome da Rua', 'Bairro',
           'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda',
           'Unidade de Medida', 'Bandeira', 'Ativo', 'Status do Valor de Venda',
           'Obs', 'Numero Rua']
merge_df.drop(labels=colunas, axis=1, inplace=True)
print(merge_df.info())
```

```python
# Remover a linhas duplicadas

merge_df.drop_duplicates(inplace=True)
display(merge_df.head(100))
```

```python
# Agrupar e contar quantos postos tem na cidade...
tabela_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
tabela_df.reset_index(inplace=True)
tabela_df.drop(labels='CNPJ da Revenda', axis=1, inplace=True)
tabela_df.rename(columns={"Revenda": "Número de Postos"}, inplace=True)
display(tabela_df)
```

```python
tabela_df['Habitantes por Postos'] = tabela_df['NumHabitantes2021'] / tabela_df['Número de Postos']
display(tabela_df.round(2))
```

```python
from pyparsing.helpers import lineStart
# Graficos
import matplotlib.pyplot as plt
plt.hist(combustiveis_df['Valor de Venda'])

# Título no gráfico
plt.title("Preço dos combustíveis - Nov/2021")

# Eixo X Rótulo
plt.xlabel("Preço (em reais)")

# Eixo Y Rótulo
plt.ylabel("Quantidade de Coletas")

# Traçar a linha vermelha tracejada com o preço médio
plt.axvline(combustiveis_df['Valor de Venda'].mean(), color='red', linestyle='dashed', linewidth=2)


# "Plota" o gráfico
plt.show()
```

```python
# Segundo Gráfico
c_mean = combustiveis_df['Valor de Venda'].groupby(by=combustiveis_df['Produto']).mean()
display(c_mean)
```

```python
import seaborn as sns

# Definido a área do gráfico
plt.figure(figsize=(10,5))
# Plotar gráfico
grafico = c_mean.plot(
    kind="barh",
    xlabel="Tipo de Combustível",
    ylabel="Preço reais/litro",
    title="Média de Preços por Combustível",
    color=["green", "red", "gray", "blue", "black", "yellow"],
    alpha=0.9
)
grafico.set_xlabel("Tipo de Combustível")
grafico.set_ylabel("Preço reais/litro")

# Grid
plt.grid()

# Remove as bordas do gráfico
sns.despine()

# Exibe
plt.show()
```

```python
from inspect import CORO_CREATED
# Personalizando a planilha Excel...
from openpyxl import load_workbook
from openpyxl.styles import PatternFill, Font, Color, Alignment

# Vai abrir o Excel no openpyxl
wb = load_workbook(excel)

# Pegar a planilha certa ... usando o Sheet Name(nome da planilha)
ws = wb['Sumário'] #ws work sheet --> planilha atual

# Pitando o cabeçalho da tabela de 'cinza'
cinza = PatternFill("solid", fgColor="cccccc")
coords = ['A1','B1']
for coord in coords:
  ws[coord].fill = cinza

# Onde o preço do  combustível for maior ou igual a 6,5 reais pinta a fonte de vermelho e negrito
# se for menor verde e negrito?
max_row = ws.max_row
num_linha = 2 
while (num_linha < max_row):
  coord = 'B'+str(num_linha)
  if ws[coord].value >= 6.5:
    ws[coord].font = Font(bold=True, color="FF0000")
  else:
    ws[coord].font = Font(bold=True, color="00FF00")
  num_linha += 1
  
# Salvar as aterações
wb.save(excel)
```
